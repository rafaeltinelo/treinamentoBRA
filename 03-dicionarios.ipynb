# Dicionários

Uma coleção de pares de chave-valor.

```python
cadastro = {'nome': 'Alexandre', 'idade': 17, 'gosta_de_python': True}
```

```python
cadastro
```

```python
type(cadastro)
```

```python
cadastro.keys()
```

```python
cadastro.values()
```

```python
cadastro.items()
```

Acessando um elemento

```python
cadastro['gosta_de_python']
```

Adicionando um novo elemento

```python
cadastro['notas'] = [10, 7, 8, 5]
```

```python
cadastro['materias'] = ['mat', 'bio', 'hist', 'port']
```

```python
cadastro
```

Iterado sobre um dicionário

```python
for chave, valor in cadastro.items():
    print(chave, valor)
```

```python
for chave, valor in cadastro.items():
    print(f'{chave}: {valor}')
```

```python
for chave, valor in cadastro.items():
    print(f'{chave.title()}: {valor}')
```

Removendo um elemento

```python
del cadastro['notas']
```
